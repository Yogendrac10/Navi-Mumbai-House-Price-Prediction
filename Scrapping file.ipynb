```python
import time
```

```python
bhk = []
near_places = []
total_area = []
prices = []
bedrooms = []


for j in range(1, 600):
    
    time.sleep(5)
    
    
    # Scorlling Slowly 
    for x in range(1, 7):
        driver.execute_script("window.scrollTo(0, window.scrollY + 2000)")
        time.sleep(1)
    
    # loading Page
    page = driver.find_element_by_xpath('//*[@id="app"]/div/div/div[3]/div[2]/div[2]')
    
    # Extracting Locations
    location = page.find_elements_by_xpath('//*[@id="srp_tuple_property_title"]')
    for i in location:
        bhk.append(i.text)
        
    # Extracting Near Places
    nearplaces = page.find_elements_by_class_name('SliderTagsAndChips__tagCount')
    for i in nearplaces:
        near_places.append(i.text)
        
    # Extracting Area    
    area = page.find_elements_by_id('srp_tuple_primary_area')
    for i in area:
        total_area.append(i.text)
        
     
    # Extracting Price
    price = page.find_elements_by_id('srp_tuple_price')
    for i in price:
        prices.append(i.text)
    
    # Extracting bedroom
    bedroom = page.find_elements_by_id('srp_tuple_bedroom')
    for i in bedroom:
        bedrooms.append(i.text)
    
    # Pagination
    if j == 1:
        try :
            next_page = driver.find_element_by_xpath('//*[@id="app"]/div/div/div[3]/div[2]/div[3]/div[3]/a')
            next_page.click()
            
        except:  
            driver.find_element_by_xpath('//*[@id="app"]/div/div/section[3]/div/div[2]/i').click()
            next_page = driver.find_element_by_xpath('//*[@id="app"]/div/div/div[3]/div[2]/div[3]/div[3]/a')
            next_page.click()
            
            
    else:
        try :
            next_page = driver.find_element_by_xpath('//*[@id="app"]/div/div/div[3]/div[2]/div[3]/div[4]/a')
            next_page.click()
            
        except ElementClickInterceptedException:  
            driver.find_element_by_xpath('//*[@id="app"]/div/div/section[3]/div/div[2]/i').click()
            next_page = driver.find_element_by_xpath('//*[@id="app"]/div/div/div[3]/div[2]/div[3]/div[4]/a')
            next_page.click()
            
    
    
```
